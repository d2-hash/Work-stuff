```python
import pandas as pd
import numpy as np
import plotly.express as px
import folium

df=pd.read_csv('/content/drive/MyDrive/output1.csv')


```

```python
df
```

```python
fig = px.density_mapbox(df, lat='Lat', lon='Lon',
                        mapbox_style="stamen-terrain")

fig
```

```python
center_lat, center_lon = df['Lat'][0], df['Lon'][0]
map_obj = folium.Map(location=[center_lat, center_lon], zoom_start=10)


for index, row in df.iterrows():
    lat, lon = row['Lat'], row['Lon']
    comparative_value = row['Comparative']
    str_value = row['STR']


    color = 'blue' if comparative_value >= 0.5 else 'red'


    popup_text = f"STR: {str_value:.2f}"


    folium.Marker([lat, lon], popup=popup_text, icon=folium.Icon(color=color)).add_to(map_obj)


map_obj
```

```python
latlist = []
lonlist = []
for row in df['Lat']:
  latlist.append(row)

for row in df['Lon']:
  lonlist.append(row)
lonlist

latav = np.mean(latlist)
lonav = np.mean(lonlist)
print('Ideal shop location is located around ' + str(latav) + ', ' + str(lonav))

```
